```python
#proses memilih kolom apa saja yang akan digunakan, nah disini saya menggunakan semua kolom kecuali kolom id dan unnamed
dfu = df5.drop(columns=['id','Unnamed: 32'])  #Data yang sudah di hapus kolom id dan unnamed
dfu.head()
```


Langkah 2 - Lakukan proses encoding pada kolom "diagnosis".

```python
le2 = LabelEncoder() # membuat objek dari LabelEncoder
#melakukan proses encoding untuk kolom diagnosis
dfu['diagnosis'] = le2.fit_transform(dfu['diagnosis']) # proses encoding

dfu.head()
```

Langkah 3 - Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
numerical = dfu.columns[dfu.columns !='diagnosis'] #berisi semua data kecuali kolom diagnosis karena tidak memiliki nilai numerik
std1 = StandardScaler()
dfu[numerical] = std1.fit_transform(dfu[numerical])
dfu.head()
```

Langkah 4 -
Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan data lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.

dfu_train, dfu_unsee = train_test_split(dfu, test_size=0.2, random_state=0, stratify=dfu['diagnosis'])
dfu_val, dfu_test = train_test_split(dfu_unsee, test_size=0.5, random_state=0, stratify=dfu_unsee['diagnosis'])
# Cek masing-masing ukuran data

print(f'Jumlah data asli: {dfu.shape[0]}')
print(f'Jumlah data train: {dfu_train.shape[0]}')
print(f'Jumlah data unsee: {dfu_unsee.shape[0]}')
print(f'Jumlah data val: {dfu_val.shape[0]}')
print(f'Jumlah data test: {dfu_test.shape[0]}')

# Cek rasio tiap label
print('=========')
print(f'Jumlah label data asli:\n{dfu["diagnosis"].value_counts()}')
print(f'Jumlah label data train:\n{dfu_train["diagnosis"].value_counts()}')
print(f'Jumlah label data val:\n{dfu_val["diagnosis"].value_counts()}')
print(f'Jumlah label data test:\n{dfu_test["diagnosis"].value_counts()}')
```
